```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Sample population dataset (World Bank - Sample)
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "Population": [1352642280, 1366417754, 1380004385, 1393409038, 1417173173]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display dataset
print(df)

# BAR CHART
plt.figure()
plt.bar(df["Year"], df["Population"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("India Population Growth (Bar Chart)")
plt.show()

# HISTOGRAM
plt.figure()
plt.hist(df["Population"], bins=5)
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Population Distribution (Histogram)")
plt.show()
```
